```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(rc={"figure.figsize":(17, 10)})
```

```python
main_path = r"C:\Users\Vadim\Desktop\СПбПУ\7_сем\УПОС\data.xlsx"
save_fig_path = r"C:\Users\Vadim\Desktop\СПбПУ\7_сем\УПОС\\"
dash_line = (5, (10, 3))
```

# 7.1.1

```python
# Read data
r1 = pd.read_excel(main_path, sheet_name="7,1,1", usecols="B,C", header=2, dtype=float)\
    .dropna().\
    rename(columns={"U=, mV": "U=",
                    "Uc, mV": "Uc"})
r2 = pd.read_excel(main_path, sheet_name="7,1,1", usecols="E,F", header=2, dtype=float)\
    .dropna().\
    rename(columns={"U=, mV.1": "U=",
                    "Uc, mV.1": "Uc"})
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(r1["Uc"], r1["U="], label="r1", linestyle="solid")
ax.plot(r2["Uc"], r2["U="], label="r2", linestyle=dash_line)

ax.legend(["r1", "r2"], prop={"size":15})
# ax.set_title("R1 and R2", fontsize=15)
ax.set_xlabel("Uc, mV", fontsize=15)
ax.set_ylabel("U=, mV", fontsize=15)
plt.savefig(save_fig_path + "7.1.1.png")
plt.show()
```

# 7.7

```python
# Read data
block_1 = pd.read_excel(main_path, sheet_name="7,7", usecols="B,C", header=2, dtype=float)
block_2 = pd.read_excel(main_path, sheet_name="7,7", usecols="E,F", header=2, dtype=float).rename(columns={"Uc, mV.1": "Uc, mV",
                                                                                                      "UΩ, mV.1": "UΩ, mV"})
block_3 = pd.read_excel(main_path, sheet_name="7,7", usecols="H,I", header=2, dtype=float).rename(columns={"Uc, mV.2": "Uc, mV",
                                                                                                      "UΩ, mV.2": "UΩ, mV"})
datasets = [block_1, block_2, block_3]
for data in datasets:
    data["K"] = data["UΩ, mV"] / (data["Uc, mV"] * 0.3)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(block_1["Uc, mV"], block_1["K"], linestyle="solid")
ax.legend(["block 1"], prop={"size":15})
ax.set_title("f = 1, KHz", fontsize=15)
ax.set_xlabel("Uc, mV", fontsize=15)
ax.set_ylabel("K", fontsize=15)
plt.savefig(save_fig_path + "7.7.1.png")
plt.show()

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(block_2["Uc, mV"], block_2["K"], linestyle="solid")
ax.legend(["block 2"], prop={"size":15})
ax.set_title("f = 1, KHz", fontsize=15)
ax.set_xlabel("Uc, mV", fontsize=15)
ax.set_ylabel("K", fontsize=15)
plt.savefig(save_fig_path + "7.7.2.png")
plt.show()

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(block_3["Uc, mV"], block_3["K"], linestyle="solid")
ax.legend(["block 3"], prop={"size":15})
ax.set_title("f = 1, KHz", fontsize=15)
ax.set_xlabel("Uc, mV", fontsize=15)
ax.set_ylabel("K", fontsize=15)
plt.savefig(save_fig_path + "7.7.3.png")
plt.show()
```
