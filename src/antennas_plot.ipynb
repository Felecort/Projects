```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
k = 2 * np.pi / 0.428
psi = np.pi / 2
```

```python

def get_all(theta):
    return (np.cos(0.428 * np.cos(theta)) - np.cos(0.428)) / (1 - np.cos(0.428))

```

```python
theta_arr = np.linspace(0, np.pi * 2, 30)
val = get_all(theta_arr)
```

```python
plt.polar(theta_arr, val);
```
